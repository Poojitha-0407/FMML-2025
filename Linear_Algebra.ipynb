**Matrix Power**

```python
 a = np.random.random([3,3])
 # positive powers of matrix
 a_1 = np.linalg.matrix_power(a, 1)
 a_9 = np.linalg.matrix_power(a, 9)
 # inverse of matrix
 a_inv_2 = np.linalg.matrix_power(a, -2)
 a_inv_4 = np.linalg.matrix_power(a, -4)
 print('matrix \n', a)
 print('\n matrix to the power 1\n', a_1)
 print('\n matrix to the power 9\n', a_9)
 print('\n matrix inverse \n', a_inv_2)
 print('\n matrix cubic inverse \n', a_inv_4)
```

**Eigenvalues and Eigenvectors**

**Solving System of Equations**

```python
# Coefficient Matrix
a = np.random.randint(1,20,[4,4])
# Dependent variable vector
b = np.array([4,9,12,7])
# solution
x = np.linalg.solve(a,b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable vector')
print(b)
print('\nSolution')
print(x)
```

```python
# Check for correctness
B = a.dot(x)
print(B)
```

**Singular Value Decomposition**

```python
from skimage import data
# download a sample image
image = data.rocket()
print(image.dtype, image.min(), image.max(), image.shape)
plt.imshow(image)
plt.show()
```

```python
# normalize image
img = image/255.0
# reorder the axes to have colour channels as the first dimension
img = np.transpose(img, axes=(2,0,1))
```

```python
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
```

```python
# S matrix should have dimensions suitable for matrix multiplication
Sigma = np.zeros((3,427,640))
for i in range(3):
  np.fill_diagonal(Sigma[i,:,:], S[i,:])
print(Sigma.shape)
```

```python
# Reconstruct the original image without any dimension reduction.
reconst = U @ Sigma @ V
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

```python
# NumPy SVD reconstruction

k = 20
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

**Dimensionality reduction**

```python
k = 5
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
